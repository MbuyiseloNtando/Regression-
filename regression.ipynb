```python
import pandas as pd
```

```python
#loading data 
df = pd.read_csv("https://raw.githubusercontent.com/artinmohajeri/Car-Data-Analisys-Pandas/main/cars.csv " )
df
```

```python
df.info()
```

Cleaning the data frame

```python
df.isna().sum()
```

```python
#Dropping null entries 
df.dropna(inplace = True)
```

```python
df.info()
df. head() 
```

```python
#Removing $, commas and any spaces from numerical columns with object entries 
dollar_cols = [] 
comma_cols = [] 
space_cols = [] 
for col in df.columns:
    if df[col].dtype=="object":
        if df[col].str.contains("$").any():
            dollar_cols.append(col)
        if df[col].str.contains(",").any():
            comma_cols.append(col)
        if df[col].str.contains("Horsepower").any():
            space_cols.append(col)
        
for col in dollar_cols:
    df[col]=df[col].str.replace("$","")
for col in comma_cols:
    df[col] =df[col].str.replace(",", "") 
for col in space_cols:
    df[col] = df[col].str.replace("Horsepower"," 150") 
```

```python
df. tail() 
```

```python
#Converting from object to numerical data types
df = df.astype({'MSRP':'int', 'Invoice':'int', 'Horsepower':'int'}) 
```

Creating the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#spilitting data and target
x= df.drop(['MSRP'], axis=1)
y=df['MSRP'] 
```

```python
#train =80, test=20
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= 0.2, random_state=0)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred= model.predict(x_test)
y_pred
```

```python
print ("The model intercept is :", model.intercept_)
print (model.coef_)
```

```python
compare= pd.DataFrame({'Actual': y_test, 'predict': y_pred})
compare. head() 
```

```python
from sklearn import metrics
import numpy as np 
```

```python
#Error evaluation 
print ('Mean Absolute Error:', metrics.mean_absolute_error (y_test, y_pred))
print ('Mean Squard Error:', metrics.mean_squared_error (y_test, y_pred))
print ('Root Mean Squard Error:', np.sqrt (metrics.mean_squared_error (y_test, y_pred)))
print ('R2 Score:', metrics.r2_score(y_test, y_pred)) 
```

```python
sns.regplot(x=y_pred, y=y_test)
plt.xlabel ('Predicted MSRP')
plt.ylabel ('Actual MSRP')
plt.title('Actual VS predicted price')
plt.show
```
